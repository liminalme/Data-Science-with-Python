```python
columns = list(df_train.describe().columns)  # save the column names as a list
```

```python
columns
```

```python
features= columns[:-1]    # define the features in the train dataset
```

```python
features
```

```python
target= columns[-1]      # define the target dataset
```

```python
target
```

```python
df_train= df_train.fillna(0)        # Fill the 
```

```python
df_train.head()
```

```python
X= df_train[features]
```

```python
X.head()
```

```python
y= df_train[[target]]
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model = model.fit(X, y)
```

```python
model
```

```python
y_pred= model.predict(X)
```

```python
np.std(y- y_pred)
```
